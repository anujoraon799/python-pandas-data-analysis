### Question 6
- Write a function dominant_species_per_island() that:
    - For each island, finds the species with the highest number of observations (non-NA species only).
    - If there's a tie, return all tied species.
    - Return a dataframe with columns: island, species, count

```python
print(penguins['species'].value_counts())
```

```python
print(
    penguins.groupby(['island', 'species'])
    .size()
)
```

```python
def dominant_species_per_island(df):
    species_count = df.groupby(['island', 'species']).size().reset_index(name='count')

    max_count = species_count.groupby('island')['count'].transform('max')

    result = species_count[species_count['count'] == max_count]

    return result

dominant_df = dominant_species_per_island(penguins)

print(dominant_df)
```

#### Q6 Insights 
###### The analysis shows that different penguin species are more commonly found in specific islands.
###### Gentoo penguins were observed the most in Biscoe island, while Chinstrap penguins were more common in Dream island.
###### Torgersen island mainly contained Adelie penguins, showing a clear difference in species distribution across islands.

### Question 7
We've prepared a synthetic dataset for you simulating an index driven by random growth rates.
- Your task is to reconstruct the index using only the starting value and the given growth rates.

```python
np.random.seed(42)
index_data = pd.DataFrame({
    "date": pd.date_range(start="2000-01-01", periods=24, freq="MS"),
    "growth_rate": np.random.uniform(-0.02, 0.02, 24),
    "index": np.nan
})
index_data.iloc[0, index_data.columns.get_loc("index")] = 100
```

```python
index_data.head()
```

```python
index_data.iloc[0]
```

```python
for i in range(1, len(index_data)):
    previous_value = index_data.loc[i - 1, 'index']
    growth = index_data.loc[i, 'growth_rate']

    index_data.loc[i, 'index'] = previous_value * (1 + growth)

index_data.head()
```

#### Q7 Insights
###### The index started with a base value of 100 and changed month by month using the given growth rates.
###### Positive growth rates increased the index value, while negative growth rates caused small decreases.
###### Each new index value was calculated using the previous month’s index value.
###### This helped recreate the complete index series from the starting value and growth data.
